```python
def co2_emission(b):
    co2 = b[b['Indicator Name']=='CO2 emissions from gaseous fuel consumption (% of total)']
    
    co2_emission = co2[co2['Country Name'].isin(['Japan','Spain','Sweden','China','Singapore','Italy','Canada'])]
    

    bar_width = 0.1

    r1 = np.arange(len(co2_emission))
    r2 = [x + bar_width for x in r1]
    r3 = [x + bar_width for x in r2]
    r4 = [x + bar_width for x in r3]
    r5 = [x + bar_width for x in r4]
    r6 = [x + bar_width for x in r5]
    r7 = [x + bar_width for x in r6]


    plt.subplots(figsize=(15, 8))
    plt.bar(r1, co2_emission['1995'],color='palegreen',width=bar_width,edgecolor='black', label='1991')
    plt.bar(r2, co2_emission['1999'],color='springgreen', width=bar_width,edgecolor='black', label='1995')
    plt.bar(r3, co2_emission['2000'],color='lime',width=bar_width,edgecolor='black', label='2000')
    plt.bar(r4, co2_emission['2003'],color='lawngreen',width=bar_width,edgecolor='black', label='2002')
    plt.bar(r5, co2_emission['2005'],color='limegreen',width=bar_width,edgecolor='black', label='2007')
    plt.bar(r6, co2_emission['2007'],color='forestgreen',width=bar_width,edgecolor='black', label='2010')
    plt.bar(r7, co2_emission['2009'],color='darkgreen',width=bar_width,edgecolor='black', label='2012')


    plt.xticks([r + bar_width*2 for r in range(len(co2_emission))], co2_emission['Country Name'])
    plt.legend()
    plt.show()
    
    return 
```

```python
co2_emission(dataset)
```

```python
def population_line_plot(c):
    
    population_growth = c[c['Indicator Name']=='Population growth (annual %)']
    
    population = population_growth[population_growth['Country Name'].isin(['Japan','Spain','Sweden','China','Singapore','Italy','Canada'])]
    
    Trans= population.transpose()
    Trans.rename(columns=Trans.iloc[0], inplace = True)
    population_transpose=Trans.iloc[4:]
    population_transpose.fillna(0,inplace=True)
    
    plt.figure(figsize=(22,10))
    plt.plot(population_transpose.index,population_transpose['Japan'],linestyle='dashed',label='Japan')
    plt.plot(population_transpose.index,population_transpose['Spain'],linestyle='dashed',label='Spain')
    plt.plot(population_transpose.index,population_transpose['China'],linestyle='dashed',label='China')
    plt.plot(population_transpose.index,population_transpose['Singapore'],linestyle='dashed',label='Singapore')
    plt.plot(population_transpose.index,population_transpose['Italy'],linestyle='dashed',label='Italy')
    plt.plot(population_transpose.index,population_transpose['Canada'],linestyle='dashed',label='Canada')
    
    plt.xlim('2000','2012')
    plt.xlabel('Year')
    plt.ylabel('%')
    plt.legend()
    plt.show()
    
    return
```

```python
population_line_plot(dataset)
```

```python
def heatmap_spain(x):
    
    indicator=['CO2 emissions from gaseous fuel consumption (% of total)',
               'Other greenhouse gas emissions (% change from 1990)',
               'Population growth (annual %)',
               'Forest area (% of land area)',
               'Agricultural land (% of land area)']
    
    spa=x.loc[x['Country Name']=='Spain']
    
    spain = spa[spa['Indicator Name'].isin(indicator)]
    
    spain_df = spain.pivot_table(spain,columns= x['Indicator Name'])
    
    spain_df.corr()
    
    plt.figure(figsize=(12,8))
    sns.heatmap(spain_df.corr(),fmt='.2g',annot=True,cmap='magma',linecolor='black')
    plt.show()
    
    return
```

```python
heatmap_spain(dataset)
```
